```python
#Modulen med funktionerna
#modulen heter hyper.py

import numpy as np

def cas(a):
    return((np.exp(a)+np.exp(-a))/2.)

def san(a):
    return((np.exp(a)-np.exp(-a))/2.)
```

Om nu ovanstående fil/modul är tillgänglig för python under namnet `hyper.py` så fungerar nedanstående som `main.py`.

```python
#Anrop för att plotta

import hyper as hp

import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-2, 2, 100)
y1=hp.cas(x)
y2=hp.san(x)

#Nu ritar vi!
plt.plot(x,y1)
plt.plot(x,y2)

plt.savefig('plot.png')
plt.show()
```
