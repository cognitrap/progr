```python
from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt
import numpy as np

# Samplingsfrekvens, sample rate: datapunkter per sekund
n = 5000
# Signalen varar mellan -d och d
d = 3


def Box():
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    y = np.heaviside(x+1,0.5)-np.heaviside(x-1,0.5)
    return(x,y)

x,y = Box()

# L책dan
plt.axis([-4,4,-1,2])
plt.scatter(x,y)
plt.show()

# Transformen
yf = fftshift(fft(y)) 
xf = fftshift(fftfreq(n*(2*d),1/n)) 

# Realdelen, scatterplot
plt.figure(figsize=(9, 4))
plt.axis([-10,10,-12000,12000])
plt.scatter(xf,yf.real)  
plt.show()


# Realdelen, linje mellan datapunkter
plt.figure(figsize=(9, 4))
plt.axis([-10,10,-12000,12000])
plt.plot(xf,yf.real)  
plt.show()
```
