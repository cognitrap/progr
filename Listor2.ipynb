## range()


Först ett begrepp som skapar list-liknande ting och som ofta används i samband med listor. range() är en så kallad inbyggd funktion (färdig funktion).

Kommandot har följande struktur: `range(start, stop, step)`
Men ofta används inte alla 3 parametrarna. Parametrarna start och step behöver inte anges men stop måste alltid anges. Skriver man `range(8)` så tolkas det automatiskt som  stop är 8 att intervallet är från 0 till 7 eftersom det är exklusivt stop, precis som vid listor.

Kommandot range används ofta då man bara vill ha en enkel upprepning:

```python
for i in range(7):
    print(i,i*3)
```

Smidigare än att arbeta med en lista.

Låt oss prova lite varianter. Och undersöka vad det är för typ av objekt egentligen.

`range(0,8)` är samma sak som `range(8)`

Vi undersöker och skriver in steg

```python
for i in range(0,7,2):
    print(i,i*3)
```

```python
for i in range(0,7,8): #steg större än intervallet
    print(i,i*3)
```

Bara att experimentera vidare.

Kommandot `range()` är inte en lista, man kan inte skriva

```python
print(range(3))
```

och få ut en lista.

Man behöver inte stega framåt.

```python
for i in range(0,-5,-1):
    print(i)
```

## Ta bort och lägga till element

Vi börjar med en lista och tar bort ett element.

```python
L=[3,4,5]
print(L)
L[1:2]=[] #dvs element med index 1, som är talet 4
print(L)
```

Vi lägger till en lista sist

```python
L=[3,4,5]
L[3:]=[1,1,0]
print(L)
```

Lägga till först

```python
L=[3,4,5]
L[:0]=["a","b","c"]
print(L)
```

Man börjar undra?
Kan man stoppa in en bit lista i listan?

```python
L=[3,4,5]
L[1:2]=["a","b","c"]
print(L)
```

## Metoder på listor

Vi introducerar metoder i python. De har formen `name.action`.
Vi ser på några exempel.

```python
L=[2,3,5]
L.append(2) #en metod
print(L)
```

Fungerade utmärkt. Kanske kan man få för sig att följande skulle fungera, det känns lite logiskt.

```python
L=[2,3,5]
Lny=L.append(2)
print(Lny)
```

Men som du ser så händer inget. Det finns ingen lista med namnet Lny fastän vi gjort en tilldelning. Denna typ av operationer kallas för 'in place' därför att den direkt gör något med objektet.

Däremot fungerar, naturligtvis

```python
L=[2,3,2,4,2]
L_nr2=L.count(2) #räknar antalet 2:or i listan
L_nr2
```

Ta bort en 2:a (det blir den första)

```python
L=[2,3,2,4,2]
L.remove(2)
L
```

Ändra listans ordning.

```python
L=[1,2,37]
L.reverse()
L
```

## Mängdbyggare

Inom matematiken finns något som kallas mängdbyggare.

$\{x|x\in\mathbb{R}\}$ som är mängden av alla reella tal (läses: mängden av $x$ sådana att , $x$ tillhör mängden $\mathbb{R}$ ) ($|$ betyder sådana att)

$\{n^{2}|n\in\mathbb{N}\}$ som är mängden av kvadraten på alla naturliga tal

Eller för diskreta ändliga mängder

$M=\{1,2,3,4\}$

$L=\{n^{2}|n\in M\}$ som är mängden $\{1, 4, 9, 16\}$

```python
M=[1,2,3,4]
L=[n**2 for n in M]
print(L)
```

Snabbt och bekvämt för att skapa listor. Fenomenet kallas för 'list comprehension'.

Vi skriver ett kort program.

```python
L=list(range(1,100,2)) #endast udda tal
Lkub=[x**3 for x in L] #list comprehension
print("Kuben på udda tal", Lkub)
```

## Slumplistor

Vi genererar listor med slumpmässigt valda naturliga tal.
Det centrala kommandot är `random`.

```python
import random

L=random.sample(range(0,10),8)
print(L)
```

Vi får tal i intervallet 0 (inkl) till 10 (exkl). Vi får 8 tal. I detta kommando kan antalet genererade slumptal inte överstiga antalet tal i intervallet. Vi kan maximalt få 10 slumpade tal, alla olika. Slump utan återläggning.
Några fler exempel.

```python
L=random.sample(range(0,10),10)
print(L)
```

```python
L=random.sample(range(0,10),11) #fungerar ej
print(L)
```
