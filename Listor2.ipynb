## range()


Först ett begrepp som skapar list-liknande ting och som ofta används i samband med listor. range() är en så kallad inbyggd funktion (färdig funktion).

Kommandot har följande struktur: `range(start, stop, step)`
Men ofta används inte alla 3 parametrarna. Parametern start och step behöver inte anges men stop måste alltid anges. Skriver man `range(8)` så tolkas det automatiskt som  stop är 8 att intervallet är från 0 till 7 eftersom det är exklusivt stop, precis som vid listor.

Kommandot range används ofta då man bara vill ha en enkel upprepning:

```python
for i in range(7):
    print(i,i*3)
```

Smidigare än att arbeta med en lista.

Låt oss prova lite varianter. Och undersöka vad det är för typ av objekt egentligen.

`range(0,8)` är samma sak som `range(8)`

Vi undersöker och skriver in steg

```python
for i in range(0,7,2):
    print(i,i*3)
```

```python
for i in range(0,7,8): #steg större än intervallet
    print(i,i*3)
```

Bara att experimentera vidare.

Kommandot `range()` är inte en lista, man kan inte skriva

```python
print(range(3))
```

och få ut en lista.

Man behöver inte stega framåt.

```python
for i in range(0,-5,-1):
    print(i)
```

## Listor 2

Vi skapar en lista med hjälp av `range()`

```python
L=list(range(8))
L
```

Vi behöver kommandot `list()` som gör om `range()` till en lista.

Föregående exempel innehåller en omvandling av en inbyggd funktion till en lista. Detta går att göra med fler datatyper. En populär är

```python
a="6" #som är en sträng
a=int(a) #omvandlar sträng till datatypen heltal
print(3*a) #nu kan vi räkna med den
```

Den klarar även knepigare saker som att tolka decimalpunkt

```python
a="6.2" #som är en sträng
a=float(a) #omvandlar sträng till datatypen flyttal
print(3*a) #nu kan vi räkna med den
```

Man börjar undra... Vad händer om vi gör

```python
a=True; b=False
a=int(a); b=int(b) 
print(a,b)
```

Strängar kan plockas i delar till en lista

```python
a="Det var länge se'n"
b=list(a)
print(b)
```

Eller om man vill ta bort något felstavat

```python
a="Konkatanering" #Vår felstavade sträng
a=list(a) #Vi gör om den till en lista
a[6]="e" #Vi rättar
a="".join(a) #Sätter ihop till sträng igen
print(a) 
```

Kommandot `"".join` lägger ihop elementen i listan och placerar inget mellan elementen. Det som är mellan citattecknen placeras mellan elementen men vi har inget mellan dem. Testa gärna med att skriva in något annat mellan citattecknen tex. `"?".join(a)`

Vi skivar upp listor på lite olika sätt för att vänja oss vid tänkandet.
Först skriver vi ner ett tankestöd i form av en tabell.




| 0 | 1 | 2 | 3 | 4 | 5 | ... | -5 | -4 | -3 | -2 | -1 |
|----|----|----|----|----|----|----|----|----|----|----|----|


där 0 är första elementet i listan och -1 är sista elementet i listan.

Om listan innehåller 5 element har vi 

| 0  | 1  | 2  | 3  | 4  |
|----|----|----|----|----|
| -5 | -4 | -3 | -2 | -1 |

Och om vi väljer listan `L=[3,7,4,1,8]` så har vi 

| 3 | 7 | 4 | 1 | 8 |
|----|----|----|----|----|
| 0  | 1  | 2  | 3  | 4  |
| -5 | -4 | -3 | -2 | -1 |

```python
L=[3,7,4,1,8]
L[2:4] # 2 inkl till 4 exkl
```

```python
L=[3,7,4,1,8]
L[2:5] # 5 är utanför sista index som är 4
```

```python
L[-2]
```

```python
L[:-2] #Startar i 0 inkl går till -2 exkl
```

```python
L[-2:] #Startar i -2 går till slutet
```

Listkommandon anger inte alltid om du ger kommandon utanför indexets gränser.
I vårt exempel har vi icke-negativa index från 0 tom. 4. Men man kan ange index 5 för att få med det sista elementet. Vad händer om man skriver

```python
L[5]
```

Lägg märke till skillnaden. `L[2:5]` ger inte felmeddelande men `L[5]` ger felmeddelande!

Så om vi kanske tänkt helt fel och skriver

```python
L[0:3000]
```

Så får vi inget felmeddelande.

Vad händer om båda gränserna anges fel (utanför listans index)?

```python
L[10:100]
```

Jo, vi får en tom lista, ok.

## Ta bort och lägga till element

Vi börjar med en lista och tar bort ett element.

```python
L=[3,4,5]
print(L)
L[1:2]=[] #dvs element med index 1, som är talet 4
print(L)
```

Vi lägger till en lista sist

```python
L=[3,4,5]
L[3:]=[1,1,0]
print(L)
```

Lägga till först

```python
L=[3,4,5]
L[:0]=["a","b","c"]
print(L)
```

Man börjar undra?
Kan man stoppa in en bit lista i listan?

```python
L=[3,4,5]
L[1:2]=["a","b","c"]
print(L)
```

Tydligen ersatte vi talet 4 med vår lista över strängar. Kanske kan vi få in vår bit utan att ta bort något? Vi försöker med

```python
L=[3,4,5]
L[1:1]=["a","b","c"]
print(L)
```
