# Mer om strängar

Strängar kan också adresseras som listor, utan att använda `list()`. Man kan dock inte ändra i dem men väl plocka ut en del av strängen.

```python
S="Ett äpple om dagen"
S[0]
```

```python
S[0:3]
```

```python
Ny=S[0:4]+"päron "+S[10:13]+S[13:18]
print(Ny)
```

Man kan dock inte ändra i strängen på något enkelt sätt.

```python
S="Ett äpple om dagen"
S[0]="A" #kommer inte ändra E till A
```
