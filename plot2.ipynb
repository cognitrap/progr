# Att rita grafer (plot)

## Mallen

För att kunna rita grafer måste vi importera moduler.
Vi har konstaterat att vi i grundutförandet inte kan göra `sin(1.2)`. Men om vi importerade math-modulen så gick det bra.
Men för att kunna rita önskar vi oss mer. För att kunna rita för hand behöver vi en värdetabell och det behöver python också. Så vi måste ha en lista med x-värden tex. `Lx=[0.1,0.2,0.3,0.4]`. Detta kan tex. `range()` hjälpa oss med. Men sedan vill vi beräkna sinusvärden för alla talen i listan. Då kan man göra en loop med `sin(Lx[i])`.
Men man har fixat detta problem och lagt det i en modul så man kan direkt skriva sinus av en lista. Kommandot `sin(Lx)` är tillåtet och python gör en ny tabell med alla sinusvärdena. Vi behöver en modul som heter numpy (numerical python).
Vi testar detta kommando först innan vi behandlar hur värdetabellen ska ritas.

```python
import numpy as np

Lx=[0.1,0.2,0.3,0.4]
Ly=np.sin(Lx)
print(Lx,Ly)
```

Man har gjort det ännu smidigare genom 2 kommandon som man infört. De heter `linspace()` respektive `arange()`. Vi använder `linspace()`. Skrivsättet blir väldigt suggestivt.

```python
import numpy as np

x=np.linspace(-2*np.pi,2*np.pi,10)
print(x)
```

Kommandot `linspace()` ska ha start och stop och antal värden `linspace(start, stop, num)`. I detta fall har jag använt intervallet från $-2\pi$ till $2\pi$. Allra sist står antalet (`num`) tal som jag vill ha. I detta fall 10 stycken.

Observera att stop ingår, se nedan.

```python
import numpy as np

x=np.linspace(-2,2,10)
print(x)
```

Vi gör en värdetabell, både x och y

```python
import numpy as np

x=np.linspace(-2*np.pi,2*np.pi,10)
print("x-värden",x)
y=np.sin(x) #nästan samma skrivsätt som i matematiken
print("y-värden",y)
```

Nu har vi vår värdetabell klar. Nu behöver vi ladda en modul som sköter om själva plottandet.

Modulen som behöver importeras, heter `matplotlib.pyplot`.
De nya kommandona framgår av koden.

```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-2*np.pi, 2*np.pi, 100)
y=np.sin(x)

#Nu ritar vi!
plt.plot(x,y)
plt.show()
```

Python justerar själv koordinatsystemet om man inte säger något, dock finns det kommando för det också.

Om du skrivit raderna i en textfil, tex. på repl.it måste du även ange kommandot `plt.show()` annars får du ingen graf att själv titta på. (jupyter klarar sig utan den, men lägger till en rad i utskriften om den inte är med)

Följden av kommandon (import, x=, y=, plot...) används ofta som en mall, sedan är det bara att fylla i och ändra efter behov.

De fyra linjerna som bildar en rektangel och ramar in själva grafen kallas för _spines_. De kan du flytta runt, ändra färg på eller göra osynliga.

## Utsmyckning

Naturligtvis kan man själv ändra koordinatsystem, rita flera kurvor i samma diagram, ändra färger på kurvorna, lägga in förklarande text osv.

Exempel följer.

```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-2*np.pi, 2*np.pi, 20)
y=np.sin(x)

#Rita
plt.plot(x,y, color="green",
         linestyle='dashed', 
         marker='o', 
         markerfacecolor='blue', 
         markersize=12, 
         linewidth=2)

plt.title('En sinusfunktion')

plt.show()

```

* `color="green"`betyder att kurvan ritas med grön färg
* `linestyle="dashed"` betyder att kurvan är streckad
* `marker="o"` innebär att datapunkterna markeras med en cirkelskiva
* `markerfacecolor="blue"` gör att cirklarna är fyllda med blått (inte gröna som kurvan)
* `markersize=12` anger storleken på markören
* `linewidth=6` anger bredden på kurvan

Experimentera. Observera stavningen av färg, det är color inte colour. Observera också att det är strängar som ska anges så citattecken måste vara med.

På repl.it finns det möjlighet att spara figuren genom att klicka på en ikon i figuren. Men man kan också skriva till ett kommando för det `plt.savefig()`.

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(-2*np.pi, 2*np.pi, 200)
y=np.sin(x)
plt.plot(x,y)

plt.savefig('plot.png')
plt.show()
```

På repl.it syns grafen till höger (i standardinställningen) och en fil med namnet plot.png skapas och läggs bland filerna till vänster där main.py ligger. Denna kan du sedan ladda ner. Flera olika format fungerar.

## Två kurvor i samma graf

Vi har ritat endast en kurva i grafen, nu ritar vi 2.

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(0, 5, 50)
y=60*np.sin(x)
plt.plot(x,y)

x=np.linspace(0, 5, 50)
y=np.exp(x)
plt.plot(x,y)

#Ritas i samma graf!

plt.show()
```

Vad är det som gör att det blir två kurvor i en graf och inte två grafer med en kurva i varje?
Det som är signifikant är att kommandot plt.show() kommer efter att vi gett två kommandon plt.plot().
Om vi ger kommandot plt.show() omedelbart efter ett plt.plot() så får vi två grafer.

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(0, 5, 50)
y=60*np.sin(x)
plt.plot(x,y)
plt.show()

x=np.linspace(0, 5, 50)
y=np.exp(x)
plt.plot(x,y)
plt.show()
```

## Staplade grafer

Rita flera grafer sidan om varandra.

För att kunna rita så kallade subplots behövs en styrning av var graferna ska placeras. Man tänker i rader och kolumner. Kommandot är `subplot(nrows, ncols, index)`.
`nrows` anger antalet rader och `ncols` anger antalet kolumner.
Index anger var grafen ska ritas. Index går från vänster till höger och uppifrån och ner.

```python
import matplotlib.pyplot as plt
import numpy as np


""" Rita båda funktionerna i två olika grafer samtidigt

subplot(nrows, ncols, index)

"""

x=np.linspace(-2*np.pi, 2*np.pi, 200)
y=np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,y)


x=np.linspace(-2*np.pi, 2*np.pi, 200)
y=np.sin(3*x)

plt.subplot(2,1,2)
plt.plot(x,y)


plt.savefig('plot2.png')
#plt.show()
```

## Teckenförklaring (legend)

Om flera kurvor ritas kan det vara bra att ge information om dem. I detta fall låter vi python bestämma var det finns plats.
Kommandona som används är:

`plot(x,y, label='abs(x)')` , `plot(x,y,'k:', label='exp(-x)*sin(x)')
` Dvs. vi anger en `label()`.

Teckenförklaringen: `legend(loc='best', fontsize='small')`

I programmet finns också ett LaTeX alternativ. Ändra med #-tecknet så de blir avkommenterade och kommentera de andra plot-anropen.

```python
#import lite annorlunda denna gång
from numpy import *
from matplotlib.pyplot import *

#abs(x)
x=linspace(-5, 5, 100)
y=abs(x)
plot(x,y, label='abs(x)')

#LaTeX alternativ
#plot(x,y, label=r'$\left|x\right|$')


#exp(-x)*sin(x)
x=linspace(-1, 5, 100)
y=exp(-x)*sin(x)
plot(x,y,'k:', label='exp(-x)*sin(x)')

#LaTeX alternativ
#plot(x,y,'k:', label=r'$e^{-x}\cdot\sin(x)$')


legend(loc='best', fontsize='medium')

show()
#savefig('plot.png')
```

## Latex kod

Symboler givna med latex-kod kan skrivas in i diagram. Vi har sett exempel nyligen, nu kommer det mer.

```python
import numpy as np
import matplotlib.pyplot as plt

#Värdetabell, obs arange()
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*np.sqrt(t))

plt.plot(t,s)

plt.axhline(y=0.5, xmin=0.0, xmax=0.9) #Horisontell linje

#Och så lite text skriven i Latex

plt.title(r'$\mathcal{A}\mathrm{sin}(2 \pi\sqrt{t})=0.5$', fontsize=20)
plt.text(0.8, -0.75, r'$\int 0.5-\mathcal{A}\mathrm{sin}(2 \pi\sqrt{t}) \mathrm{dt}$', fontsize=15)
plt.text(1.1, 0.05, r'$\mathcal{A}\mathrm{sin}(2 \pi\sqrt{t})$',fontsize=15)

#Markering på axlarna
plt.xlabel('time (s)')
plt.ylabel('volt (mV)')
plt.show()
```

Vi tittar på raderna för Latex.

`plt.title(r'$\mathcal{A}\mathrm{sin}(2 \pi\sqrt{t})=0.5$', fontsize=20)`

plt.title() är som tidigare.
Direkt efter vänsterparentesen kommer ett `r` som anger att det är raw-data som anges. Vad som är raw anges sedan med ' så det finns en avslutande också. Nu kommer latex på vanligt sätt. \\$ för att det är matematik så kallat _inline_. `\pi` ger den grekiska bokstaven $\pi$ (som jag i denna text också har skrivit på detta sättet).

Nytt är också markeringar på axlarna:

`plt.xlabel('time (s)')`
`plt.ylabel('volt (mV)')`

Som alternativ användes `arange(start, stop, step)` i stället för `linspace(start, stop, num)`

Dvs. i arange anges steglängden (step) i stället för antalet element (num).

stop i linspace() ingår i elementen som produceras.
stop i arange() ingår inte


```python
#från och med 0 till 4, inte 5
t = np.arange(0, 5, 1)
print(t)
```

### Skolkoordinatsystem

'spines' är namnet på de långa linjer som går mellan markörer. Förinställt är 4 stycken som bildar en låda runt grafen, se tidigare grafer. Dessa kan flyttas och göras osynliga.

I koden nedan utförs:

- Vänstra spine flyttas till origo. Högra spine blir osynlig, färg sätts till `none`.

- Nedersta spine flyttas till origo och översta görs osynlig.

```python
import numpy as np
import matplotlib.pyplot as plt


#fig = plt.figure()

x = np.linspace(-np.pi, np.pi, 100)
y = 2*np.sin(x)

ax = plt.gca() #Säger bara att vi vill arbeta med gällande bilds axlar: get current axes/chart

ax.plot(x, y)
ax.grid(True)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

plt.xlim(-np.pi,np.pi)

#Lägger in pilarna
plt.arrow(3.14, -0.006, 0.1, 0, width=0.007, color="k", clip_on=False, head_width=0.12, head_length=0.12)
plt.arrow(0.016, 2.04, 0, 0.1, width=0.007, color="k", clip_on=False, head_width=0.12, head_length=0.12)




plt.savefig("sin.png")
plt.show()
```

De fyra första argumenten i kommandot för att rita en pil `plt.arrow(3.14, -0.006, 0.1, 0)` är x-koordinat, y-koordinat, längd i x-led, längd i y-led.

Vi ritar enbart en pil

```python
import numpy as np
import matplotlib.pyplot as plt
plt.arrow(0.2, 0.1, 0.2, 0.4, width=0.015, color="k", clip_on=False, head_width=0.12, head_length=0.12)
plt.show()
```
