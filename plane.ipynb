### Rät linje i 2D och 3D samt plan i 3D




Rät linje i 2 dimensioner

```python
import numpy as np
import matplotlib.pyplot as plt

#Parameter
t=np.linspace(0, 5, 50)

#Parameterframsttällning (x(t),y(t))
x=2*t
y=3*t

plt.plot(x,y)
plt.show()
```

En rät linje i 3 dimensioner

```python
# importavdelningen
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

#i egen ruta, ta bort på repl.it
%matplotlib tk

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# parametrarna t
t=np.linspace(1, 20, 1000)

# linje
x = 2+t
y = -3+t
z= 1+t

# plotta
plt.plot(x,y,z)

# visa plotten
plt.show()
```

Ett plan

```python
# importavdelningen
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TkAgg')
import numpy as np


#i egen ruta, ta bort på repl.it
#%matplotlib tk

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# parametrarna t och s
tl=np.linspace(-20, 20, 1000)
sl=np.linspace(-20, 20, 1000)

# plane
s,t=np.meshgrid(sl,tl) #Kartesiska produkten
x = 2+t-s
y = -3+t+2*s
z= 1+t+s

# plotta
ax.plot_surface(x,y,z,linewidth=0, antialiased=False, cmap="coolwarm")

# visa plotten
plt.show()
```

### Vektorer

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,11)
y = np.linspace(0,1,11)

#Lägger i princip noll-vektorer i alla punkter
u = np.zeros((11,11))
v = np.zeros((11,11))

#Men vi vill ha en vektor som inte är en noll-vektor
u[5,5] = 0.2
v[5,5] = 0.4

plt.axis('equal') #axlar lika stora

#scale=1 behövs för annars gör python automatisk skalning
plt.quiver(x, y, u, v, scale=1, units='xy')
```

Om man vill ha flera vektorer anges de med listor. Första elementet i varje lista är för vektor 1, andra elementet för vektor 2 osv. I koden ritas vektorer med utgångslägena (0,0), (0.2,0.3) och (0.2,0.1). Riktningen anges på motsvarande sätt. Riktningen för vektorn i punkten (0,0) ges av (0.2,0.1)

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

#start
x0=[0., 0.2, 0.2]
y0=[0., 0.3, 0.1]
dx=[0.2, 0.2, 0.3]
dy=[0.1, 0.1, 0.1]

plt.axis('equal') #axlar lika stora
ax.axis([-0.2,0.6,-0.2,0.6])


#scale=något tex 1; behövs för annars gör python automatisk skalning
plt.quiver(x0, y0, dx, dy, scale=1, units='xy')
```

Vi styr vektorn med en grafisk kontroll (widget)

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact



def f(x00,x01,y00,y01,dx0,dx1,dy0,dy1):
#start
    fig, ax = plt.subplots()

    x0=[x00, x01]
    y0=[y00, y01]
    dx=[dx0, dx1]
    dy=[dy0, dy1]

    plt.axis('equal') #axlar lika stora
    ax.axis([-10,10,-10,10])

#scale=något tex 1; behövs för annars gör python automatisk skalning
    plt.quiver(x0, y0, dx, dy, scale=1, units='xy')
    return

interact(f, x00=(-5,5), x01=(-5,5),
        y00=(-5,5), y01=(-5,5),
        dx0=(-5,5), dx1=(-5,5),
        dy0=(-5,5), dy1=(-5,5))



```

Vi ritar 2 vektorer i 3D.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets



def f(x00,x01,y00,y01,z00,z01,dx0,dx1,dy0,dy1,dz0,dz1):
#start   
    %matplotlib tk

    fig = plt.figure()
    ax = fig.gca(projection='3d')

    x0=[x00, x01]
    y0=[y00, y01]
    z0=[z00, z01]
    dx=[dx0, dx1]
    dy=[dy0, dy1]
    dz=[dz0, dz1]

    ax.axis([-5,5,-5,5])
    ax.set_zlim(bottom=None, top=5)
    ax.quiver(x0, y0, z0, dx, dy, dz)
    plt.show()

    return

interact(f, x00=(-5,5), x01=(-5,5),
        y00=(0,5), y01=(0,5),
        z00=(0,5), z01=(0,5),
        dx0=widgets.IntSlider(min=-5, max=5, step=1, value=4), 
        dx1=widgets.IntSlider(min=-5, max=5, step=1, value=-2),
        dy0=widgets.IntSlider(min=-5, max=5, step=1, value=-4),
        dy1=widgets.IntSlider(min=-5, max=5, step=1, value=-1),
        dz0=widgets.IntSlider(min=-5, max=5, step=1, value=2), 
        dz1=widgets.IntSlider(min=-5, max=5, step=1, value=1))
```
