### Rät linje i 2D och 3D samt plan i 3D




Rät linje i 2 dimensioner

```python
import numpy as np
import matplotlib.pyplot as plt

#Parameter
t=np.linspace(0, 5, 50)

#(x(t),y(t))
x=2*t
y=3*t

plt.plot(x,y)
plt.show()
```

En rät linje i 3 dimensioner

```python
# importavdelningen
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

#i egen ruta, ta bort på repl.it
%matplotlib tk

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# parametrarna t
t=np.linspace(1, 20, 1000)

# linje
x = 2+t
y = -3+t
z= 1+t

# plotta
plt.plot(x,y,z)

# visa plotten
plt.show()
```

Ett plan

```python
# importavdelningen
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np


#i egen ruta, ta bort på repl.it
%matplotlib tk

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# parametrarna t och s
tl=np.linspace(-20, 20, 1000)
sl=np.linspace(-20, 20, 1000)

# plane
s,t=np.meshgrid(sl,tl) #Kartesiska produkten
x = 2+t-s
y = -3+t+2*s
z= 1+t+s

# plotta
ax.plot_surface(x,y,z,linewidth=0, antialiased=False, cmap="coolwarm")

# visa plotten
plt.show()
```
