# Att göra val

Ofta hamnar man i en situation där man måste välja vad som ska göras. Tex. om en variabel är större än 0 så ska programmet göra på ett sätt men om variabeln är mindre än 0 så ska programmet göra något annat. Detta hanteras med så kallade 'if-satser'.

I övre delen av figuren illustreras ett vanligt linjärt flöde där satserna följer på varandra uppifrån och ner. Inga alternativ.

I mitten och nederst illustreras en punkt i programmet där vi ska utföra olika satser beroende på värdet på en variabel.

I flödet har jag skrivit in en rad som ska göras då det är sant att `x>0` och det är satsen `x=x+1`. 

Observera att denna sats har ett indrag (indent). Det är inte av estetiska skäl utan det _måste_ vara så. Det är en del av språkets syntax. Detta kan ibland ställa till problem om man kopierar och klistrar in kod; indragen går ibland förlorade.

Observera tecknet kolon dvs :
Det _måste_ vara där.

Man kan också skriva in flera satser. Python tolkar alla indragna satser som ett block eller svit av kod; båda uttrycken används i svenskan. Se nästa cell.

```python
x=4 #Ändra och prova
if x>0:
    x=x+1 #dessa 3 rader är ett block eller en svit
    d=7
    print("x var större än 0")
else:
    x=x-1 #dessa 3 rader är ett block eller en svit
    d=2*x
    print(d)
print(x)
```

Det finns också en enklare konstruktion.

```python
x=int(input("Ange ett heltal"))

if x>0:
    print(x)
print("denna skrivs alltid ut")
```

Anger du tex. 3 så skrivs 3 ut. Anger du tex. -5 så skrivs inte x ut. Däremot skrivs den sista (inte indragen) printsatsen _alltid_ ut.

Indragen i python måste vara minst ett mellanslag. Varje block av kod måste ha samma indrag. Men olika block kan ha olika indrag. Dock är seden den att man i regel använder 4 mellanslag och ALDRIG har olika indrag i en kod; alltid heltalsmultiplar av 4.
