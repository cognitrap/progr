# Att göra val

Ofta hamnar man i en situation där man måste välja vad som ska göras. Tex. om en variabel är större än 0 så ska programmet göra på ett sätt men om variabeln är mindre än 0 så ska vi programmet göra något annat. Detta hanteras med så kallade 'if-satser'.

I flödet har jag skrivit in en rad som ska göras då det är sant att `x>0` och det är satsen `x=x+1`. 

Observera att denna sats har ett indrag (indent). Det är inte av estetiska skäl utan det _måste_ vara så. Det är en del av språkets syntax. Detta kan ibland ställa till problem om man kopierar och klistrar in kod; indragen går ibland förlorade.

Observera tecknet kolon dvs :
Det _måste_ vara där.

Man kan också skriva in flera satser. Python tolkar alla indragna satser som ett block av kod. Se nästa cell.

```python
x=4 #Ändra och prova
if x>0:
    x=x+1 #dessa 3 rader är ett block
    d=7
    print("x var större än 0")
else:
    x=x-1 #dessa 3 rader är ett block
    d=2*x
    print(d)
print(x)
```

Det finns också en enklare konstruktion.

```python
x=int(input("Ange ett heltal")) #Förklaras i nästa avsnitt

if x>0:
    print(x)
print("denna skrivs alltid ut")
```

Anger du tex. 3 så skrivs 3 ut. Anger du tex. -5 så skrivs inte x ut. Däremot skrivs den sista (inte indragen) printsatsen _alltid_ ut.
