### Vektorfält

Att rita vektorfält görs i huvudsak med kommandot `quiver` som betyder pilkorg eller koger.

Nedan ritas vektorfältet $ \overline{F}=\left(1,-1\right) $.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

#Vectorfield
u = 1
v = -1

plt.quiver(x,y,u,v)
plt.show()
```
