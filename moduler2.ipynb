### Moduler och Namnrum (Namespaces)

När man importerar en modul så har den sitt eget namnrum, sin egen uppsättning av namn. Dessa namn kan komma i konflikt med varandra. `sin()` kan betyda olika saker beroende på vilka moduler som är importerade.

Moduler som är vanliga är: math, numpy, scipy, matplotlib, panda.

Man kan ha
`math.sin()`
`numpy.sin()`
`scipy.sin()`

Tex. tar math bara enkla tal medan numpy tar listor.

För att ta reda på vilka kommandon som finns i modulen scipy skriver man `dir(scipy)`. Man måste först ha gjort en import av modulen.

```python
import scipy
dir(scipy) #Rätt många kommandon
```

Om man inte laddar någon modul så laddas ändå alltid modulen builtins. Du ska inte ladda den själv det görs automatiskt då python startar.

```python
dir(__builtins__) #Två understreck på varje sida
```

Man kan kontrollera om ett viss kommando finns i en viss modul. Om den finns svarar python med True.

```python
'abs' in dir(__builtins__)
```

```python
'sin' in dir(__builtins__)
```

```python
import math

'sin' in dir(math) #Observera inga understreck här
```
