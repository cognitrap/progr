Derivata definieras med hjälp av en differenskvot. I regel som
$$ \frac{f\left(x+h\right)-f\left(x\right)}{\left(x+h\right)-x}= \frac{f\left(x+h\right)-f\left(x\right)}{h}$$

och denna kan användas på diskreta data. Vid programmering, tex. när vi ritar grafer så finns inga kontinuerliga funktioner utan det finns värdetabeller med diskreta data tex. skapade med kommandon som `linspace()`.

I fysiken där läge mäts som funktion av tid ges hastighet som ändring i läge dividerat med ändring i tid, en differnskvot. När man sedan skall beräkna acceleration så är det ändring i hastighet dividerat med ändring i tid, ytterligare en differens. Efter två differens-kvoter ser data inte så bra ut längre.

Differenskvoten ovan kallas Newtons differenskvot. En, i numeriska sammanhang, bättre differenskvot är den symmetriska differenskvoten som använder en punkt före och en efter den punkt som vi beräknar kvoten för

$$\frac{f\left(x+h\right)-f\left(x-h\right)}{\left(x+h\right)-\left(x-h\right)}=\frac{f\left(x+h\right)-f\left(x-h\right)}{2h}$$

Sedan beräknar vi den Newtonska differenskvoten och drar linjesegment mellan punkterna.

```python
import matplotlib.pyplot as plt
import random as r
import numpy as np

x=np.linspace(-10,10,20)
#Vi måste arbeta med listor
x=list(x)
noise=[]
y=[]
for i in range(0,20):
    noise.append(r.random())
for i in range(0,20):
    y.append(2*x[i]+5*noise[i])
    
#Newtons differenskvot
yN=20*[0]
for i in range(0,19):
    yN[i]=(y[i+1]-y[i])/(x[i+1]-x[i])

#Symmetrisk differenskvot
yS=20*[0]
for i in range(1,19):
    yS[i]=(y[i+1]-y[i-1])/(x[i+1]-x[i-1])

plt.plot(x,y,'o')
plt.plot(x,yN,'--')
plt.plot(x,yS,'-')
```

Även om effekten inte är slående så ser man tydligt att den symmetriska differenskvoten alltid har variationer som är mindre än närliggande variationer för Newtons differenskvot. Och naturligtvis kan tankegången utvidgas till att blanda in ytterligare punkter för beräkning av differenskvoten som approximation till derivata i en praktisk situation.
