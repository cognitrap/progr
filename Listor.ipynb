# Listor

En väldigt användbar datatyp är listor. Förståelse för den hjälper en att även förstå andra datatyper som vi kommer till.

## Ange en lista

Vi anger en lista och ger den ett nam:

```python
L=[1,2,2,5,3,7]
```

Listan heter M och är en lista av tal. Talen kan vara lika, tex. 2, men ses som olika. I mängder kan man inte ha mer än ett tal 2.

Hur kommer vi åt talen i listan? Jo med något som påminner om index. Index skrivs i matematiken som $L_3$, där talet 3 är ett index-värde, men vid användning av datorer, som är radorienterade får man markera index på annat sätt.

Observera att första talet har index 0, inte index 1!

Låt oss kalla på första talet i listan, dvs. talet 1. Observera att du måste ha kört kod-snutten där L definieras, annars får du felmeddelande.

```python
L[0]
```

Och det sista talet i listan

```python
L[5]
```

En lista med 6 tal har index från 0 till 5.

Men det är lite mer än så. Listorna har vad man skulle kunna kalla för ett baklänges-index också. Om första talet har index 0 och andra talet index 1; vad händer om man skriver L[-1]? Jo man får då sista talet, talet 7!

```python
L[-1]
```

Det går att fortsätta med negativa tal

```python
L[-2]
```

Kan man gå flera varv? Är kanske L[6] samma som L[0], typ moduloräkning? Det är bara att pröva, python talar om ifall du inte följer syntaxen (språkets grammatik)

```python
L[6]
```

Förmodligen accepterar den L[-6] men inte L[-7]

```python
L[-6]
```

```python
L[-7]
```

Ok, vi fattar.

## Del av lista

När vi anger att index ska gå mellan 2 värden i matematiken kan man ange det som $L_i$ då $i=2...5$

I python skriver man med kolon inom hakparenteserna

```python
L[0:2]
```

Men det blev bara 2 tal fastän vi angett från 0 till 2, vilket är 0 och 1 och 2.

Python tolkar det som från och med 0 till 2. Dvs. inklusive 0 men exklusive 2. Argument för detta finns i dokumentationen av python. Man kan se det som L[inkl:exkl]

Denna typ av 'utsnitt' kallas på engelska för 'slice'.

Ok, det kan man acceptera. Men vad händer då om man skriver L[0:0] där den ena nollan är inklusive och den andra exklusive? Testa är en ok metod.

```python
L[0:0]
```

Python svarar [] vilket betyder den tomma mängden, inga tal från listan är med, således exklusive vinner.

En lista behöver inte ha samma typer av element utan de kan blandas. En lista kan innehålla heltal, booleska variabler, strängar osv.

```python
M=[2,3,-1,True,'w']
M[2:5]
```

Om man ska ha en lista från ett visst element till slutet räcker det med att ange starten tex. M[2:]

```python
M[2:]
```

Och om vi vänder på det M[:2]?

```python
M[:2]
```

Vill man ha reda på hur många element en lista innehåller så använder man len(M)

```python
len(M)
```

Om längden är 5 så går index från 0 till 4, inte 1 till 5.
