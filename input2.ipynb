### Att läsa filer med data

Ibland är datamängden stor och arbetet med att skriva in den i listor för att rita en graf är stor. Python kan läsa olika filer med data.

```python
#Filens namn anges först, sedan w för att vi ska skriva och + 
#för att den ska skapas om den inte finns.
#Om filen redan finns och är skriven till gäller append.


#Öppnar filen för arbete och ger den ett enklare namn.
f= open("datafil1.txt","w+")

#%d skriv heltal, \r return, \n newline

for i in range(10):
     #f.write("This is line %d\r\n" % (i+1))
     f.write("This is line {}\n".format((i+1)))


#Vi stänger filen när vi är färdiga
f.close()




#För att leka lite med filen så öppnar vi den igen. a append
f=open("datafil1.txt", "a")
f.write("Nu var det slut")
f.close()


#Vi skriver ut filen. Det görs genom att öppna den för läsning och sedan läsa. r read.
#Filen läses automatiskt rad för rad
with open('datafil1.txt', 'r') as f:
    print(f.read())

#Man behöver ingen close om man använder with :-)
#f.close()
```

????????Vi skriver mätdata till en fil. I detta exempel måste vi för illustrationens skull göra det själva. I en mätsituation kan filen vara skapad av ett annat program och lagrats eller exporterats till .txt eller .csv

```python
#Ska läsa en fil
f=open("datafil2.txt", "r")

#Läs hela filen och skriv ut den, ingen while!
content =f.read()
print("Hela filen direkt\n", content)

f.close()


#Man kan också läsa radvis på annat sätt än med while.
#Vi öppnar filen igen, läser då automatiskt från början

f=open("datafil2.txt", "r")

#Lägg märke till att filen f1 är en lista med varje rad som ett element i listan.
#Kommandot är också tydligt readlines(). Obs inget while men ändå rader.
f1=f.readlines()
print("rad 1:", f1[0])
print("rad 2:", f1[1])
print("rad 3:", f1[2])
print("rad 4:", f1[3])

f.close()
```

Vi ska nu öppna och läsa en fil med ändelsen (extension) csv.
(file extension översätts till svenska som filsuffix eller filändelse)
csv betyder comma separated values. En fil där data separeras med ett kommatecken (,). Talen använder decimalpunkt så betydelserna hålls isär. Detta är en sak man måste se upp med då man importerar från olika program.
https://fileinfo.com/extension/csv

Först en bild på den datafil som vi ska läsa.

![csv-fil](Pictures/pictdatacsv.png)

```python
#DENNA SKA BORT
#Vi behöver en modul för att läsa .csv (en ren textfil i detta fall)
import csv

with open('datacsv.txt','r') as f:
  data = csv.reader(f)
  for row in data:
        print(row)
#Automatiskt close. Så nästa gång vi använder with börjar vi från början
#i filen


import matplotlib.pyplot as plt
import numpy as np


#Nu ska vi läsa in dessa listor
#Med rader ungefär som ovan läses varje rad som en lista
#Men första elementet i raden är x och det andra y, så de måste
#sorteras i nya listor

x=[]; y=[]
with open('datacsv.txt','r') as f:
  data = csv.reader(f)
  for row in data:
        x.append(int(row[0]))
        y.append(int(row[1]))

print("x",x)
print("y",y)





plt.axis([0,6,0,30])

plt.plot(x, y, 'ro--')
#red circle

plt.savefig('plot.png')
plt.show()
```
