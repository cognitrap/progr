```python
# importavdelningen
import matplotlib.pyplot as plt
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D

# lite bekv√§mlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# variabeln/parametern t
t=np.linspace(1, 20, 1000)

# kurvan
x = t * np.sin(2 * t)
y = t * np.cos(2 * t)
z = t


# plotta
ax.plot(x,y,z,'-b')

# visa plotten
plt.show()
```
