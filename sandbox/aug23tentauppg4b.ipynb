```python
from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt
import numpy as np

# Antal samplingspunkter, sample rate: datapunkter per sekund
n = 5000
# Signalens varaktighet r frn -d till d
d = 3

# Olika funktioner att studera
# Genererar en sinusvg med given frekvens
def gen_wave1(frekv):
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    y = np.sin(2*np.pi*frekv*x)
    return(x,y)

# Genererar en gauss-kurva med en given varians
def gen_wave2(A):
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    y = 1/np.sqrt(4*np.pi*A)*np.exp(-x**2/4/A)
    return(x,y)


# Genererar en kurva 
def gen_wave3():
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
   # y = 1/np.sqrt(4*np.pi*A)*np.exp(-x**2/4/A)
    y=np.exp(-abs(x))*np.cos(x)
    return(x,y)




# Genererar en sinusvg genom anrop
# Frekvensen r satt till 100 Hz
#x,y = gen_wave1(100)

# Genererar en gaussisk funktion genom anrop
# Variansen A r satt till 0,0001
#x,y=gen_wave2(0.0001)

x,y=gen_wave3()

# Ritar funktionen som ska transformeras
plt.figure(figsize=(9, 4))  # x respektive y i tum.
plt.axis([-6,6,-1,1])
plt.plot(x,y)
plt.show()


# Kommandot fr att f ut fouriertransformen fr en av funktionerna
yf = fftshift(fft(y)) # Sjlva transformen
xf = fftshift(fftfreq(n*(2*d),1/n)) # Skapar en vettig frekvensskala

#teori
x = np.linspace(-d,d,n*(2*d), endpoint=False)
   # y = 1/np.sqrt(4*np.pi*A)*np.exp(-x**2/4/A)
y=0.5*(2/(1+(x-1)**2)+2/(1+(x+1)**2))
plt.figure(figsize=(9, 4))  # x respektive y i tum.
plt.axis([-6,6,-1,2])
plt.plot(x,y)
plt.show()

plt.figure(figsize=(9, 4))  # x respektive y i tum.

# Vi plottar imaginrdelen av transformen (nr det r sinus)
# Fr gaussfunktionen ska man anvnda np.abs(yf)
# Ofta vill man inte se hela och i y-led behver man inte se spikens max, drfr .axis
plt.axis([-2,2,0,6000])  
#plt.scatter(xf,np.abs(yf), color="blue", marker=".")
plt.plot(xf,np.abs(yf), color="blue", marker=".")

plt.show()

```
