```python
####EJ KORREKT
### TOLKNING AV KURVORNA
from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt
import numpy as np

# Samplingsfrekvens, sample rate: datapunkter per sekund
n = 5
# Signalen varar mellan -d och d
d = 2


def Box():
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    print("x=",x)
    y = np.heaviside(x+1,0.5)-np.heaviside(x-1,0.5)
    print("y=",y)
    return(x,y)

x,y = Box()

# LÃ¥dan
plt.axis([-4,4,-1,2])
plt.plot(x,y)
plt.show()

# Transformen
yf = fftshift(fft(y)) 
xf = fftshift(fftfreq(n*(2*d),1/n)) 
print("xf=",xf)
print("yf=",yf.real)

# Realdelen
plt.axis([-2,2,-8,8])
plt.plot(xf,yf.real)  
plt.show()
```

```python
####EJ KORREKT
### TOLKNING AV KURVORNA
from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt
import numpy as np

# Samplingsfrekvens, sample rate: datapunkter per sekund
n = 11
# Signalen varar mellan -d och d
d = 2


def sinkvot():
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    #print("x=",x)
    y = np.sin(2*x)/(np.pi*x)
    print("y=",y)
    return(x,y)

x,y = sinkvot()

# sinkvot
#plt.axis([-4,4,-1,2])
plt.plot(x,y)
plt.show()

# Transformen
yf = fftshift(fft(y)) 
xf = fftshift(fftfreq(n*(2*d),1/n)) 
print("xf=",xf)
print("yf=",yf.real)

# Realdelen
#plt.axis([-2,2,-8,8])
plt.plot(xf,abs(yf))  
plt.show()
```
