# Jupyters Notebook

Denna text kan sägas bestå av två komponenter. Den ena är vanlig text, som du läser just nu. Det andra är kod skriven i python.

Koden skriven i python kan köras genom att du placerar markören i en cell med python-kod. Tryck sedan på symbolen Run. Här kommer en cell med kod nedanför (just nu är du i en cell med text).

```python
2+3
```

Testa att placera markören i kod-cellen och tryck på Run. En ny rad med resultatet ska visas precis nedanför python-koden. Kod-cellen står det In på och python skriver där det står Out.

Du kan ändra i kod-cellen. Markera bara och ändra som du vill. Du ändrar inte i originalet (det ligger på min dator).

Om du vill ta bort alla resultat-celler så gör Cell>All Output>Clear

Om du vill köra alla kod-celler så gör Cell>Run All

Om du vill köra en kod-cell så placera markören i cellen som ska köras och tryck på Run

Även text-celler behöver köras med Run. Speciellt om det finns avancerade matematiska tecken, de är då skrivna i ett layoutspråk som heter LaTeX.

En cell kan också köras genom att placera markören i cellen och sedan göra shift-return (håll nere shift och tryck på return).
