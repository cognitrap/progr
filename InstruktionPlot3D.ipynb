Nu är koden helt klar för att köras.

````python
# importavdelningen
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# variabeln/parametern t
t=np.linspace(1, 20, 1000)

# kurvan
x = t * np.sin(2 * t)
y = t * np.cos(2 * t)

# plotta
ax.plot(x,y,t,'-b')

# visa plotten
plt.show()

````

Nedan demonstreras att plotten kan göras interaktiv. Kanske känner du igen glidaren från Geogebra? Faktum är att Geogebra tidigare var skriven i Python.

````python
# importavdelningen
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# import för interactiv graf i IPython (Anaconda)
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets


# kurvan
def f(a):
    #fig = plt.figure()
    ax = plt.axes(projection='3d')
    t=np.linspace(1, 20, 1000)
    x = np.sin(a*t)
    #y = t
    y=np.cos(t)
    ax.plot(x,y,t,'-b')
    plt.show()
    return x,y

# Interagera
interactive(f, a=5)
````
