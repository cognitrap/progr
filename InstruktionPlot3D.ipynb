````python
# importavdelningen
import matplotlib.pyplot as plt
import numpy as np

# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# variabeln/parametern t
t=np.linspace(1, 20, 1000)

# kurvan
z=t
x = t * np.sin(2 * t)
y = t * np.cos(2 * t)

# plotta
ax.plot(x,y,z,'-b')

# visa plotten
plt.show()

````

Nedan demonstreras att plotten kan göras interaktiv. Kanske känner du igen glidaren från Geogebra? Faktum är att Geogebra tidigare var skriven i Python.

````python
# importavdelningen
import matplotlib.pyplot as plt
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D

# import för interactiv graf i IPython (Anaconda)
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets


# kurvan
def f(a):
    #fig = plt.figure()
    ax = plt.axes(projection='3d')
    t=np.linspace(1, 20, 1000)
    x = a*t
    y = t * np.sin(a * t)
    z = t * np.cos(a * t)
    
    ax.plot(x,y,z,'-b')
    plt.show()
    return x,y,z

# Interagera
interactive(f, a=5)
````

Vi behöver specifikt mplot3d: `from mpl_toolkits import mplot3d`
Till denna modul skickar man ett keyword-argument `projection='3d'` så erhålls axlar i 3 dimensioner.
Vi sätter samman dessa och erhåller ett 3 dimensionellt system, dock utan en graf.


````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
````

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()

''''''''''
# importing libraries 
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt 
  
# defining surface and axes 
x = np.outer(np.linspace(-2, 2, 10), np.ones(10)) 
y = x.copy().T 
z = np.cos(x ** 2 + y ** 3) 
  
fig = plt.figure() 
  
# syntax for 3-D plotting 
ax = plt.axes(projection ='3d') 
  
# syntax for plotting 
ax.plot_surface(x, y, z, cmap ='viridis', edgecolor ='green') 
ax.set_title('Surface plot geeks for geeks') 
plt.show() 
````
